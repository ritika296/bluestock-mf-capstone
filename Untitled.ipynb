# Bluestock Mutual Fund Analytics Capstone

## Project Overview

This project presents an end-to-end analytics solution for the Indian Mutual Fund industry using Python and Power BI. The project covers data cleaning, exploratory data analysis, financial performance analysis, advanced analytics, and dashboard development.

## Objectives

- Build an ETL pipeline
- Perform Exploratory Data Analysis
- Calculate financial performance metrics
- Perform advanced risk analytics
- Build an interactive Power BI dashboard
